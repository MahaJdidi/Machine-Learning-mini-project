```python
# Importing necessary libraries
import pandas as pd

# Access the uploaded file's name dynamically

# Reading the CSV file into a pandas DataFrame
df = pd.read_csv(r"C:\Users\ASUS\Downloads\Employee Attrition.csv")


# Display the first few rows of the DataFrame
df.head()
```

```python
import numpy as np
df['Attrition'] = np.random.choice(['Yes', 'No'], size=len(df), p=[0.6, 0.4])

# Display the first few rows of the DataFrame
print(df.head())
```

```python
df = df.drop(columns=['Work_accident'])
df.head()
```

```python
df_salary_dummies = pd.get_dummies(df['salary'])

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, df_salary_dummies], axis=1)

# Drop the original 'Salary' column if no longer needed
df.drop('salary', axis=1, inplace=True)

# Display the first few rows of the updated DataFrame
df.head(10)
```

```python
df_salary_dummies = pd.get_dummies(df['Attrition'])

# Concatenate the dummy variables with the original DataFrame
df = pd.concat([df, df_salary_dummies], axis=1)

# Drop the original 'Salary' column if no longer needed
df.drop('Attrition', axis=1, inplace=True)

# Display the first few rows of the updated DataFrame
df.head(10)
```

```python
df.rename(columns={
    'low': 'low salary',
    'medium': 'medium salary',
    'high': 'high salary',
    'Yes': 'Yes_attrition',
    'No': 'No_attrition'  # Ensure 'No' is in lowercase if that's how it appears in your DataFrame
}, inplace=True)
df.head()
```

```python
import pandas as pd

# Save the DataFrame to a CSV file
df.to_csv(r"C:\Users\ASUS\Downloads\Employee_Attrition.csv", index=False)
```
